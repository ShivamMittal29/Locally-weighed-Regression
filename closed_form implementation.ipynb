```python
def predict(X,theta):
    return np.dot(X,theta)
def thetaderive(X,Y):
    
    Y=np.mat(Y)
    firstpart=np.dot(X.T,X)
    Z=np.linalg.pinv(firstpart)
    secondpart=np.dot(X.T,Y)
    theta=np.dot(Z,secondpart)
    
    return theta
```
